```python
import pandas as pd
import numpy as np
df = pd.read_csv('./indexsort(1).csv')
df
```

```python
df.pop('Unnamed: 0')
```

```python
from sklearn.impute import KNNImputer

# 从原始数据帧中提取所有数值型列
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# 创建KNNImputer对象
imputer = KNNImputer(n_neighbors=5)

# 对数值型列进行填充
filled_values = imputer.fit_transform(df[numeric_cols])  # 使用KNNImputer填充数值型列

# 创建填充后的数据帧
filled_df = pd.DataFrame(filled_values, columns=numeric_cols)  # 将填充后的数组转换为dfFrame格式，列名为数值型列的列名

# 将填充后的数值型列与原始数据帧合并
df[numeric_cols] = filled_df  # 将填充后的数值型列替换原始数据帧中的对应列
df
```

```python
indexsort = df
# 找到填充后的数据帧中包含空值的行
df = df.drop(['ths_controlling_holder_held_ratio_stock','ths_the_sw_industry_index_code_stock'],axis=1)
df
```

```python
df = df.drop(['time','thscode'],axis=1)
```

```python
df = df.drop('ths_stock_short_name_stock',axis=1)
```

```python
df = df.drop('date',axis=1)
```

```python
X=df.drop('ths_chg_ratio_stock_shifted',axis=1)
y=df['ths_chg_ratio_stock_shifted']
X_train=X[indexsort['time']<'2023-10-23']
y_train=y[indexsort['time']<'2023-10-23']
X_test=X[indexsort['time']=='2023-10-23']
y_test=y[indexsort['time']=='2023-10-23']
```

```python
from sklearn.linear_model import Ridge, Lasso

lasso_model = Lasso(alpha=0.02)
lasso_model.fit(X_train,y_train)
weights_l = [lasso_model.intercept_]
weights_l.extend(lasso_model.coef_)
weights_l
```

```python
pred_lasso=lasso_model.predict(X_test)
pred_lasso
```

```python
test_set = indexsort.loc[X_test.index,['time','thscode','ths_stock_short_name_stock','ths_chg_ratio_stock']]
res = pd.concat([test_set,pd.DataFrame(pred_lasso,index=test_set.index)],axis=1)
res
```
